# Part 1. Data Understanding - Banking Marketing Dataset


### The dataset in this project was created by S. Moro, P. Rita, and P. Cortez and was obtained from UCI Machine Learning Repository. It contains information about direct marketing campaigns conducted by a Portuguese banking institution, where the authors' goal was to predict whether a client would subscribe to a term deposit (target variable y).

```python
import pandas as pd
```

### Initial preview of the dataset.

Viewing the first rows of the dataset helps confirm that the data was loaded correctly and provides an initial understanding of the variables.

```python
df = pd.read_csv('bank.csv', sep=";")
df.head()
```

### Dataset dimensions
The size of the dataset indicates how many observations (rows) and variables (columns) are available for analysis.

```python
df.shape
```

### Data types and potential missing values

Examining data types ensures variables are in appropriate formats for analysis. Checking for missing values helps identify potential data quality issues that may require preprocessing.

```python
summary = pd.DataFrame({
    "data_type":df.dtypes,
    "unique_values":df.nunique(),
    "missing_values":df.isna().sum()
})
summary
```

### Variables included in the dataset
Reviewing the list of variables helps identify the types of information available, including demographic characteristics, financial indicators, and details of a previous marketing campaign.

```python
df.columns
```


### Consistency with the metadata

The authors of the dataset has provided the metadata describing each variable, its role, it's type, and demographic. Reviewing the unique values of categorical variables helps confirm consistency between the dataset and provided metadata, and helps identify placeholders values such as "unknown".

#### Categorical Values Overview

```python
categorical_variables = df.select_dtypes(exclude=["int", "float"])

for column in categorical_variables.columns:
    print(f"\n{column}")
    print(df[column].unique())
```

Previously, when checking for missing values, the results showed that there is no missing data. However, several categorical variables contain the value 'unknown', which may represent missing or unavailable information and will be addressed during preprocessing.
